## Theoretical Questions & Answers

### 1. What are the key differences between SQL and NoSQL databases?

**Key Differences:**
- **Schema**: SQL uses fixed schema, NoSQL uses flexible/dynamic schema
- **Data Structure**: SQL uses tables and rows, NoSQL uses documents, key-value pairs, or graphs
- **Scalability**: SQL scales vertically, NoSQL scales horizontally
- **ACID Properties**: SQL guarantees ACID, NoSQL uses eventual consistency
- **Joins**: SQL uses joins, NoSQL uses denormalization or embedding
- **Querying**: SQL uses SQL language, NoSQL has various query languages
- **Transactions**: SQL supports multi-row transactions, NoSQL traditionally didn't (now some do)
- **Examples**: SQL (MySQL, PostgreSQL), NoSQL (MongoDB, Redis, Cassandra)

### 3. Explain the concept of collections in MongoDB.

**Collections in MongoDB:**
- **Definition**: Collections are groups of documents similar to tables in SQL databases
- **Dynamic Schema**: Collections don't enforce a schema; documents in a collection can have different structures
- **Document Storage**: Each collection stores JSON-like documents (BSON format)
- **Metadata**: Collections maintain metadata about themselves (indexes, validation rules, etc.)
- **Automatic Creation**: Collections are created automatically when first document is inserted
- **Structure**: Suitable for storing related data but with flexible structure
- **Example**: A "users" collection stores multiple user documents with varying fields
- **Benefits**: Allows storing complex nested data structures within a single document

### 4. How does MongoDB ensure high availability using replication?

**MongoDB Replication for High Availability:**
- **Replica Sets**: Multiple MongoDB nodes maintain copies of the same data
- **Primary Node**: Accepts all write operations and maintains the authoritative version of data
- **Secondary Nodes**: Replicate data from the primary node asynchronously
- **Automatic Failover**: If primary fails, secondary is automatically elected as new primary
- **Data Redundancy**: Ensures data is not lost if one server fails
- **Read Preferences**: Can read from secondary nodes to distribute load
- **Oplog (Operation Log)**: Records all operations on primary to sync secondaries
- **Flexibility**: Can have 3, 5, or more nodes in a replica set
- **Consistency**: Eventual consistency with options for strong consistency

### 6. What is the role of indexes in MongoDB, and how do they improve performance?

**Indexes in MongoDB:**
- **Definition**: Indexes are data structures that store a sorted copy of selected fields
- **Performance**: Enable MongoDB to scan fewer documents when executing queries
- **Query Speed**: Can dramatically reduce query execution time
- **Types of Indexes**: Single field, compound, text, geospatial, TTL, wildcard indexes
- **Primary Key Index**: MongoDB automatically creates _id index on every collection
- **Trade-off**: Trade increased storage space for faster query execution
- **Write Performance**: Indexes slow down write operations as they need to be updated
- **Index Selection**: Should be created on frequently queried fields
- **explain() Method**: Used to analyze query performance and index usage
- **Best Practices**: Create indexes based on actual query patterns

### 7. Describe the stages of the MongoDB aggregation pipeline.

**MongoDB Aggregation Pipeline Stages:**
- **$match**: Filters documents based on specified conditions
- **$project**: Reshapes documents, selects/adds/removes fields
- **$group**: Groups documents by specified field and performs aggregations
- **$sort**: Sorts documents by specified field(s)
- **$limit**: Limits the number of documents passed to next stage
- **$skip**: Skips specified number of documents
- **$lookup**: Performs left outer join with another collection
- **$unwind**: Deconstructs array fields into separate documents
- **$out**: Writes results to a new collection
- **$facet**: Returns multiple aggregations in a single stage
- **$bucket**: Groups documents into buckets based on conditions
- **Order Matters**: Pipeline processes documents stage by stage sequentially

### 8. What is sharding in MongoDB? How does it differ from replication?

**Sharding vs Replication:**
- **Sharding (Horizontal Partitioning)**: Distributes data across multiple servers based on a shard key
- **Replication (Data Redundancy)**: Maintains copies of the same data on multiple nodes
- **Purpose of Sharding**: Handle large datasets and high throughput by distributing load
- **Purpose of Replication**: Ensure data availability and fault tolerance
- **Data Distribution**: Sharding splits data into subsets; replication duplicates data
- **Scalability**: Sharding allows horizontal scaling of storage and processing
- **Availability**: Replication provides high availability and backup
- **Shard Key**: Critical choice in sharding; determines data distribution
- **Cluster Architecture**: Sharded cluster includes config servers, mongos routers, and shards
- **Use Together**: Usually used together - each shard is a replica set

### 10. What are the ACID properties in the context of MongoDB transactions?

**ACID Properties in MongoDB:**
- **Atomicity**: A transaction is all-or-nothing. Either all operations succeed or all fail
- **Consistency**: Database moves from one consistent state to another; no partial updates
- **Isolation**: Concurrent transactions don't interfere with each other; each sees consistent view
- **Durability**: Once committed, transaction data persists even after system failure
- **Multi-Document Transactions**: MongoDB 4.0+ supports transactions across multiple documents
- **Replica Set Requirement**: Transactions work within a replica set environment
- **Session Management**: Transactions are managed within sessions
- **Write Concern**: Determines durability level of writes
- **Rollback**: Failed transactions are automatically rolled back
- **Performance Trade-off**: Transactions have performance overhead compared to single operations

### 11. What is the purpose of MongoDB's explain() function?

**MongoDB explain() Function:**
- **Purpose**: Analyzes query execution and provides performance metrics
- **Usage**: `db.collection.find().explain("executionStats")`

  - `queryPlanner`: Shows query plan without executing

- **Index Usage**: Shows whether indexes are used and how effectively
- **Query Optimization**: Helps identify slow queries and optimization opportunities
- **Execution Plan**: Displays the strategy MongoDB uses to execute query
- **Stage Information**: Details about each stage (COLLSCAN, IXSCAN, FETCH, etc.)

### 13. What is the difference between a primary and a secondary node in a replica set?

**Primary vs Secondary Nodes:**
- **Primary Node**: 
  - Accepts all write operations
  - Maintains authoritative version of data
  - Only one primary per replica set
  - Replicates data to secondary nodes via oplog
- **Secondary Nodes**:
  - Replicate data from primary asynchronously
  - Refuse write operations by default
  - Can handle read operations (with read preferences)
  - Can be elected as primary if current primary fails
- **Data Consistency**: Secondaries eventually consistent with primary
- **Automatic Failover**: If primary fails, secondaries elect new primary
- **Configuration**: Replica set minimum 3 nodes (1 primary + 2 secondary)

### 15. Explain the concept of embedded documents and when they should be used.

**Embedded Documents in MongoDB:**
- **Definition**: Documents nested within other documents as subdocuments
- **Structure**: Store related data within a single document
- **Example**: Address information embedded within user document
- **Benefits**:
  - Better data locality and query performance
  - Atomic operations on entire document
  - No need for joins
  - Simpler queries
- **When to Use**:
  - One-to-one relationships
  - One-to-few relationships
  - Frequently accessed together data
  - Small related data sets
- **When NOT to Use**:
  - Large arrays that grow indefinitely
  - One-to-many relationships with many items
  - Duplicate data across multiple documents
- **Trade-offs**: Denormalization trades write efficiency for query simplicity

### 16. What is the purpose of MongoDB's $lookup stage in aggregation?

**MongoDB $lookup Stage:**
- **Purpose**: Performs left outer join with another collection
- **Similar to SQL**: LEFT JOIN in SQL databases
- **Syntax**: Joins documents from current collection with documents from another
- **Parameters**:
  - `from`: Foreign collection to join
  - `localField`: Field from current collection
  - `foreignField`: Field from foreign collection
  - `as`: Name of array to hold joined documents
- **Usage**: Combining data from multiple collections
- **Performance**: Can be slow with large datasets; consider indexing join fields
- **Result**: Returns all documents from left collection plus matched documents from right
- **Denormalization Alternative**: Often preferable to use embedded documents instead

### 19. How do MongoDB transactions differ from SQL transactions?

**MongoDB vs SQL Transactions:**
- **Scope**:
  - SQL: Transaction can span multiple tables and rows
  - MongoDB: Multi-document transactions available (4.0+), but document is atomic by default
- **Isolation Level**:
  - SQL: Offers various isolation levels (READ UNCOMMITTED, READ COMMITTED, etc.)
  - MongoDB: Snapshot isolation for multi-document transactions
- **Performance**:
  - SQL: Full ACID guarantee may impact performance
  - MongoDB: Optimized for document-level atomicity
- **Schema**:
  - SQL: Enforced schema requires normalization
  - MongoDB: Flexible schema allows embedding related data
- **Cross-Collection Transactions**: MongoDB can handle multi-document transactions across collections
- **Write Concern**: MongoDB transactions support different write concern levels
- **Rollback**: Both support automatic rollback on failure

### 20. What are the main differences between capped collections and regular collections?

**Capped vs Regular Collections:**
- **Capped Collections**:
  - Fixed size collection with automatic document removal
  - FIFO (First In First Out) behavior
  - Old documents automatically deleted when limit reached
  - Cannot exceed specified size
  - Useful for logs, cache, or limited-size queues
- **Regular Collections**:
  - Unlimited size; grows as documents are added
  - Documents removed only explicitly (delete operations)
  - More flexible for general use cases
  - Better for long-term data storage
- **Creation**:
  - Capped: `db.createCollection("logs", {capped: true, size: 1000000, max: 10000})`
  - Regular: `db.createCollection("users")`
- **Indexes**: Both support indexes
- **Performance**: Capped collections are faster for sequential writes

### 21. What is the purpose of the $match stage in MongoDB's aggregation pipeline?

**MongoDB $match Stage:**
- **Purpose**: Filters documents to pass only those matching specified conditions
- **Position**: Should be early in pipeline to reduce documents processed
- **Syntax**: `{$match: {field: value, ...}}`
- **Operators supported**: $eq, $gt, $lt, $lte, $gte, $ne, $in, $nin, $exists, etc.
- **Benefits**: 
  - Reduces number of documents in pipeline
  - Improves performance by filtering early
  - Uses indexes if available
- **Example**: `{$match: {status: "completed", amount: {$gt: 100}}}`
- **Performance**: Should place $match as early as possible in pipeline
- **Equivalent**: Similar to WHERE clause in SQL

### Dataset: Superstore Dataset

```python
# Import Required Libraries and Setup

import pandas as pd
from pymongo import MongoClient
from pymongo.errors import ConnectionFailure
import csv

# Connection details (adjust based on your MongoDB setup)
# For local MongoDB, use: mongodb://localhost:27017/
# For production, use connection string from MongoDB Atlas
connection_string = "mongodb://localhost:27017/"
database_name = "superstore_db"
collection_name = "Orders"

# Establish connection to MongoDB
try:
    client = MongoClient(connection_string)
    # Test connection
    client.admin.command('ping')
    print("MongoDB connection successful!")
except ConnectionFailure:
    print("Failed to connect to MongoDB. Make sure MongoDB is running on localhost:27017")
    
# Access database and collection
db = client[database_name]
orders_collection = db[collection_name]
```

### 1. Write a Python script to load the Superstore dataset from a CSV file into MongoDB.

```python
# Load Superstore dataset from CSV and insert into MongoDB

csv_file_path = "/workspaces/MongoDB/superstore - superstore.csv"

# Read CSV file
df = pd.read_csv(csv_file_path)

# Convert dataframe to list of dictionaries
records = df.to_dict('records')

# Clear existing data if any
orders_collection.delete_many({})

# Insert records into MongoDB
if records:
    result = orders_collection.insert_many(records)
    print(f"Successfully inserted {len(result.inserted_ids)} documents into {collection_name} collection")
    print(f"First 3 documents inserted:")
    for doc in orders_collection.find().limit(3):
        print(doc)
else:
    print("No records to insert")
```

### 2. Retrieve and print all documents from the Orders collection.

```python
# Retrieve all documents from the Orders collection

# Get all documents
all_documents = list(orders_collection.find())

print(f"Total documents in collection: {len(all_documents)}")
print("\nFirst 5 documents from the Orders collection:\n")

for i, doc in enumerate(all_documents[:5], 1):
    print(f"Document {i}:")
    for key, value in doc.items():
        print(f"  {key}: {value}")
    print()
```

### 3. Count and display the total number of documents in the Orders collection.

```python
# Count and display total number of documents

total_count = orders_collection.count_documents({})
print(f"Total number of documents in the {collection_name} collection: {total_count}")
```

### 4. Write a query to fetch all orders from the "West" region.

```python
# Fetch all orders from the West region

west_orders = list(orders_collection.find({"Region": "West"}))

print(f"Number of orders from West region: {len(west_orders)}")
print("\nFirst 5 orders from West region:")

for i, order in enumerate(west_orders[:5], 1):
    print(f"\nOrder {i}:")
    print(f"  Order ID: {order.get('Order ID')}")
    print(f"  Customer: {order.get('Customer Name')}")
    print(f"  Region: {order.get('Region')}")
    print(f"  Sales: {order.get('Sales')}")
    print(f"  Profit: {order.get('Profit')}")
```

### 5. Write a query to find orders where Sales is greater than 500.

```python
# Find orders where Sales is greater than 500

high_sales_orders = list(orders_collection.find({"Sales": {"$gt": 500}}))

print(f"Number of orders with Sales > 500: {len(high_sales_orders)}")
print("\nFirst 5 orders with Sales > 500:")

for i, order in enumerate(high_sales_orders[:5], 1):
    print(f"\nOrder {i}:")
    print(f"  Order ID: {order.get('Order ID')}")
    print(f"  Product: {order.get('Product Name')}")
    print(f"  Sales: ${order.get('Sales'):.2f}")
    print(f"  Profit: ${order.get('Profit'):.2f}")
```

### 6. Fetch the top 3 orders with the highest Profit.

```python
# Fetch top 3 orders with highest Profit

top_profit_orders = list(orders_collection.find().sort("Profit", -1).limit(3))

print("Top 3 orders with highest Profit:\n")

for i, order in enumerate(top_profit_orders, 1):
    print(f"Rank {i}:")
    print(f"  Order ID: {order.get('Order ID')}")
    print(f"  Customer: {order.get('Customer Name')}")
    print(f"  Product: {order.get('Product Name')}")
    print(f"  Sales: ${order.get('Sales'):.2f}")
    print(f"  Profit: ${order.get('Profit'):.2f}")
    print()
```

### 7. Update all orders with Ship Mode as "First Class" to "Premium Class."

```python
# Update all orders with Ship Mode "First Class" to "Premium Class"

# Count before update
first_class_before = orders_collection.count_documents({"Ship Mode": "First Class"})
print(f"Orders with 'First Class' before update: {first_class_before}")

# Update operation
result = orders_collection.update_many(
    {"Ship Mode": "First Class"},
    {"$set": {"Ship Mode": "Premium Class"}}
)

# Count after update
first_class_after = orders_collection.count_documents({"Ship Mode": "First Class"})
premium_class_after = orders_collection.count_documents({"Ship Mode": "Premium Class"})

print(f"\nUpdate completed!")
print(f"Documents modified: {result.modified_count}")
print(f"Orders with 'First Class' after update: {first_class_after}")
print(f"Orders with 'Premium Class' after update: {premium_class_after}")
```

### 8. Delete all orders where Sales is less than 50.

```python
# Delete all orders where Sales is less than 50

# Count before deletion
total_before = orders_collection.count_documents({})
low_sales_before = orders_collection.count_documents({"Sales": {"$lt": 50}})
print(f"Total documents before deletion: {total_before}")
print(f"Orders with Sales < 50 before deletion: {low_sales_before}")

# Delete operation
result = orders_collection.delete_many({"Sales": {"$lt": 50}})

# Count after deletion
total_after = orders_collection.count_documents({})
print(f"\nDeletion completed!")
print(f"Documents deleted: {result.deleted_count}")
print(f"Total documents after deletion: {total_after}")
```

### 9. Use aggregation to group orders by Region and calculate total sales per region.

```python
# Use aggregation to group by Region and calculate total sales per region

pipeline = [
    {
        "$group": {
            "_id": "$Region",
            "total_sales": {"$sum": "$Sales"},
            "total_profit": {"$sum": "$Profit"},
            "order_count": {"$sum": 1},
            "avg_sales": {"$avg": "$Sales"}
        }
    },
    {
        "$sort": {"total_sales": -1}
    }
]

results = list(orders_collection.aggregate(pipeline))

print("Sales by Region:\n")
print(f"{'Region':<15} {'Orders':<10} {'Total Sales':<15} {'Total Profit':<15} {'Avg Sales':<15}")
print("-" * 70)

for result in results:
    region = result.get("_id")
    orders = result.get("order_count")
    sales = result.get("total_sales")
    profit = result.get("total_profit")
    avg_sales = result.get("avg_sales")
    
    print(f"{region:<15} {orders:<10} ${sales:<14.2f} ${profit:<14.2f} ${avg_sales:<14.2f}")
```

### 10. Fetch all distinct values for Ship Mode from the collection.

```python
# Fetch all distinct values for Ship Mode

distinct_ship_modes = orders_collection.distinct("Ship Mode")

print("Distinct Ship Mode values:\n")
for i, mode in enumerate(sorted(distinct_ship_modes), 1):
    count = orders_collection.count_documents({"Ship Mode": mode})
    print(f"{i}. {mode} - {count} orders")
```

### 11. Count the number of orders for each category.

```python
# Count the number of orders for each category

pipeline = [
    {
        "$group": {
            "_id": "$Category",
            "order_count": {"$sum": 1},
            "total_sales": {"$sum": "$Sales"},
            "total_profit": {"$sum": "$Profit"}
        }
    },
    {
        "$sort": {"order_count": -1}
    }
]

results = list(orders_collection.aggregate(pipeline))

print("Order Count by Category:\n")
print(f"{'Category':<20} {'Order Count':<15} {'Total Sales':<15} {'Total Profit':<15}")
print("-" * 65)

for result in results:
    category = result.get("_id")
    count = result.get("order_count")
    sales = result.get("total_sales")
    profit = result.get("total_profit")
    
    print(f"{category:<20} {count:<15} ${sales:<14.2f} ${profit:<14.2f}")

print(f"\n{'Total':<20} {sum([r['order_count'] for r in results]):<15}")
```
